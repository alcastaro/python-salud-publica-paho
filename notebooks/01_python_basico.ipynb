Más información en:

[Tipos de datos](https://www.w3schools.com/python/python_datatypes.asp)

```python
# Texto
# El texto siempre se debe colocar entre comillas dobles "" o simples ''
x = "Hello World"
```

```python
# Para ver el valor de una variable se requiere llamarla con su nombre o imprimirla
x
print(x)
```

```python
# Identificando el tipo de dato
type(x)
```

```python
# Datos numéricos
# Los números enteros son integer
# Usamos la función type() para saber el tipo de dato
a = -1
type(a)
```

```python
# Los números decimales son flotantes o float
b = 1.5
type(b)
```

```python
# Los tipos booleanos admiten el valor verdadero (True) o falso (False)
c = True
type(c)
```

```python
# La ausencia de valor se asigna con la palabra None
d = None
type(d)
```

```python
# Podemos ver el valor None
print(d)
```

## Estructuras con múltiples datos

```python
# Listas
l = [1,2,3,4,5,5,5,"a"]
l 
```

```python
# Tupla
t = (1,2,3,4,5,5,5,'a')
t
```

```python
# Conjunto - Solo admite valores no repetidos
# Veremos que no se repite el cinco
set(t)
```

```python
# Diccionario
d={'uno':1,'letra':'a'}
d['letra']
```

```python
# Se pueden obtener las llaves de los diccionarios con el método keys
d.keys()
```

## Indexación de datos

```python
# Puedo seleccionar secciones o elementos de una estructura de datos utilizando []
# Siempre el primer elemento es el valor cero
print(l)
print("El primer elemento es:")
print(l[0])
```

```python
# La indexación es aplicable también a textos
# Utilizo dos puntos para establecer un rango
palabra= "Hola Mundo"
print(palabra[0:3])
```

```python
# Puedo utilizar valores negativos para especificar los valores que no deseo
# Los dos puntos significa todo hasta la penúltima posición
print(palabra[:-2])
```
